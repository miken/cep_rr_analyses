```python
# coding: utf-8
import pandas as pd
xl = pd.ExcelFile('data.xlsx')
xl.sheet_names
dsets = {}
for n in xl.sheet_names:
    dsets[n] = xl.parse(n)
```

```python
final_client_rr = pd.DataFrame(columns=["Client", "Round", "Final_RR"])
```

```python
data = dsets["14X"]
rnd = "14X"
```

```python
raw_collist = ["Client", "Round", "Final"]
data["Round"] = rnd
sliced = data[raw_collist]
```

```python
sliced
```

```python
final_client_rr.append(sliced)
```

```python

# What is the expected variation in response rates?
# First let's construct a final_rr datasets across all rounds
raw_collist = ["Client", "Round", "Final"]
final_client_rr = pd.DataFrame(columns=raw_collist)
for rnd, data in dsets.iteritems():
    data["Round"] = rnd
    sliced = data[raw_collist]
    final_client_rr = final_client_rr.append(sliced)
```

```python
final_client_rr.reset_index(drop=True)
```

```python
final_client_rr.describe()
```

```python
rr_global_stats = final_client_rr.describe()
rr_global_stats
```

```python
rr_global_stats.ix['std', 'Final']
```

```python
mean = rr_global_stats.ix['mean', 'Final'] 
std = rr_global_stats.ix['std', 'Final']
# Lowest RR range:
low_rr = mean - 3*std
high_rr = mean + 3*std
```

```python
low_rr
```

```python
high_rr
```

```python
# 95% range of response rates
low_rr_95 = mean - 3*std
high_rr_95 = mean + 3*std
print 'We can expect 95% of the times for response rates to be between {low}% and {high}%'.format(
    low=low_rr_95,
    high=high_rr_95
)
# 67% range of response rates
low_rr_67 = mean - 2*std
high_rr_67 = mean + 2*std
print 'We can expect 67% of the times for response rates to be between {low}% and {high}%'.format(
    low=low_rr_67,
    high=high_rr_67
)
```

```python
# Helper functions
def to_perc(val):
    return "{0:.0f}%".format(val*100)

# 95% range of response rates
low_rr_95 = mean - 3*std
high_rr_95 = mean + 3*std
print 'We can expect 95% of the times for response rates to be between {low}-{high}'.format(
    low=to_perc(low_rr_95),
    high=to_perc(high_rr_95)
)
# 67% range of response rates
low_rr_67 = mean - 2*std
high_rr_67 = mean + 2*std
print 'We can expect 67% of the times for response rates to be between {low}-{high}'.format(
    low=to_perc(low_rr_67),
    high=to_perc(high_rr_67)
)
```
